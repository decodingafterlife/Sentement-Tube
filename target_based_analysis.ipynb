```python
from googleapiclient.discovery import build
import json

def get_youtube_api_service(api_key):
    youtube_api_service = build('youtube', 'v3', developerKey=api_key)
    return youtube_api_service

def search_videos(api_service, keyword, max_results=10):
    search_response = api_service.search().list(
        q=keyword,
        part='id',
        type='video',
        maxResults=max_results,
        eventType='completed'  # Exclude live streams
    ).execute()

    video_ids = [item['id']['videoId'] for item in search_response['items']]
    return video_ids

def main():
    # Replace 'YOUR_API_KEY' with your actual API key
    api_key = "AIzaSyDFFVUkYavn8C3BMAssUShtVVz7NAupOlE"
    keyword = input("Enter the keyword to search for: ")

    youtube_api_service = get_youtube_api_service(api_key)
    video_ids = search_videos(youtube_api_service, keyword)

    print("Top Trending Videos for '{}':".format(keyword))
    for video_id in video_ids:
        print(video_id)

if __name__ == "__main__":
    main()

```

```python
from youtube_transcript_api import YouTubeTranscriptApi

def get_youtube_api_service(api_key):
    youtube_api_service = build('youtube', 'v3', developerKey=api_key)
    return youtube_api_service

def get_video_details(api_service, video_id):
    video_response = api_service.videos().list(
        part='snippet',
        id=video_id
    ).execute()

    if 'items' in video_response and video_response['items']:
        return video_response['items'][0]['snippet']
    else:
        return None

def get_video_transcript(video_id):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        return transcript
    except Exception as e:
        print(f"Error retrieving transcript: {str(e)}")
        return None

def main():
    # Replace 'YOUR_API_KEY' with your actual API key
    api_key = "AIzaSyDFFVUkYavn8C3BMAssUShtVVz7NAupOlE"
    video_id = input("Enter the YouTube video ID: ")

    youtube_api_service = get_youtube_api_service(api_key)
    video_details = get_video_details(youtube_api_service, video_id)

    if video_details:
        title = video_details['title']
        description = video_details['description']
        
        print(f"Title: {title}")
        print(f"Description: {description}")

        transcript = get_video_transcript(video_id)
        if transcript:
            print("\nTranscript:")
            for entry in transcript:
                print(f"{entry['start']} - {entry['start'] + entry['duration']}: {entry['text']}")
        else:
            print("Transcript not available.")
    else:
        print("Video details not found.")

if __name__ == "__main__":
    main()
```

```python
from googleapiclient.discovery import build

def get_youtube_api_service(api_key):
    youtube_api_service = build('youtube', 'v3', developerKey=api_key)
    return youtube_api_service

def get_video_captions(api_service, video_id):
    captions_response = api_service.captions().list(
        part='snippet',
        videoId=video_id
    ).execute()

    english_captions = [caption for caption in captions_response.get('items', []) if caption['snippet']['language'] == 'en']
    return english_captions

def get_caption_text(api_service, caption_id):
    caption_response = api_service.captions().download(
        id=caption_id
    ).execute()

    return caption_response.decode('utf-8')

def main():
    # Replace 'YOUR_API_KEY' with your actual API key
    api_key = "AIzaSyDFFVUkYavn8C3BMAssUShtVVz7NAupOlE"
    video_id = input("Enter the YouTube video ID: ")

    youtube_api_service = get_youtube_api_service(api_key)
    captions = get_video_captions(youtube_api_service, video_id)

    if captions:
        print("English Transcripts:")
        for caption in captions:
            caption_text = get_caption_text(youtube_api_service, caption['id'])
            print(caption_text)
    else:
        print("No English captions found for this video.")

if __name__ == "__main__":
    main()
```
